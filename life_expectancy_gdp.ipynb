```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('all_data.csv')
df.head()
```

```python
df.info()
```

## Countries

```python
df.Country.unique()
```

## Has life expectancy increased over time in the six nations

```python
sns.scatterplot(x=df.Year, y=df['Life expectancy at birth (years)'], hue=df.Country)
plt.show()
plt.clf()
```

From the scatter plot, we can tell the life expectency in western countries have increased steadily from 2000 to 2014. \
The life expectency in `Zimbabwe` has increased by a great amount from 2000 to 2014

## Has GDP increased over time in the six nations

```python
sns.scatterplot(x=df.Year, y=df['GDP'], hue=df.Country)
plt.show()
plt.clf()
```

From our analysis we can conclude that the GDP of `China` and the `USA` have seen the most significant increase \
The GDP of most of the other nations have remained the same relatively

## Is there a correlation between GDP and life expectancy of a country

```python
sns.scatterplot(x=df.GDP, y=df['Life expectancy at birth (years)'], hue=df.Country)
plt.show()
plt.clf()
```

All countries seem to have a positive growth in life expectancy regardless of their GDP

## What is the average life expectancy in these nations

```python
# getting means
chile_mean = np.mean(df['Life expectancy at birth (years)'][df.Country == 'Chile'])
china_mean = np.mean(df['Life expectancy at birth (years)'][df.Country == 'China'])
germany_mean = np.mean(df['Life expectancy at birth (years)'][df.Country == 'Germany'])
mexico_mean = np.mean(df['Life expectancy at birth (years)'][df.Country == 'Mexico'])
USA_mean = np.mean(df['Life expectancy at birth (years)'][df.Country == 'United States of America'])
zimbabwe_mean = np.mean(df['Life expectancy at birth (years)'][df.Country == 'Zimbabwe'])
```

```python
sns.barplot(x=['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe'], y=[chile_mean, china_mean, germany_mean, mexico_mean, USA_mean, zimbabwe_mean], palette='Accent')
plt.xlabel('Country')
plt.ylabel('Life expectancy at birth (years)')
plt.show()
```

We can tell that `Germany` on average has the highest life expectancy and most other countries hover around the same value \
`Zimbabwe` on the other hand has a very low life expectancy compared to the others
